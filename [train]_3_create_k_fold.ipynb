```python
import json
from sklearn.model_selection import KFold

for k in [5]:
    segmented_data = json.load(open('./data/train_data/tokenized_data/segmented_data.json', 'r'))['data']

    kf = KFold(n_splits=k)
    kf.get_n_splits(segmented_data)

    for idx, (train_index, dev_index) in enumerate(kf.split(segmented_data)):
        fold_idx = idx + 1
        json.dump({'version': f'train_{fold_idx}', 'data': [segmented_data[i] for i in train_index]},
                  open(f'./data/train_data/full_data/{k}_fold_word/train_{fold_idx}.json', 'w', encoding='utf-8'), ensure_ascii=False)

        json.dump({'version': f'dev_{fold_idx}', 'data': [segmented_data[i] for i in dev_index]},
                  open(f'./data/train_data/full_data/{k}_fold_word/dev_{fold_idx}.json', 'w', encoding='utf-8'), ensure_ascii=False)
        
    for article in segmented_data:
        title = article['title']
        for paragraph in article['paragraphs']:
            paragraph['context'] = paragraph['context'].replace('_', ' ')
            for qa in paragraph['qas']:
                qa['question'] = qa['question'].replace('_', ' ')
                qa['answers'] = [{'answer_start': answer['answer_start'], 'text': answer['text'].replace('_', ' ')} for answer in qa['answers']]
                if 'plausible_answers' in qa:
                    qa['plausible_answers'] = [{'answer_start': answer['answer_start'], 'text': answer['text'].replace('_', ' ')} for answer in qa['plausible_answers']]

    for idx, (train_index, dev_index) in enumerate(kf.split(segmented_data)):
        fold_idx = idx + 1
        json.dump({'version': f'train_{fold_idx}', 'data': [segmented_data[i] for i in train_index]},
                  open(f'./data/train_data/full_data/{k}_fold_syllable/train_{fold_idx}.json', 'w', encoding='utf-8'), ensure_ascii=False)

        json.dump({'version': f'dev_{fold_idx}', 'data': [segmented_data[i] for i in dev_index]},
                  open(f'./data/train_data/full_data/{k}_fold_syllable/dev_{fold_idx}.json', 'w', encoding='utf-8'), ensure_ascii=False)
```
